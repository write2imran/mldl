## Frensel type Sine and Cosine Integrals

```python
import numpy as np, matplotlib.pyplot as plt
```

```python
from scipy.special import fresnel
```

```python
#np.info(fresnel)
ssa, cca = fresnel(np.linspace(-4, 4, 1000))
```

```python
plt.plot(ssa, cca, 'b-');
#plt.axes().set_aspect('equal');
plt.show()
```
